```python
from transformers import AutoModelForCausalLM
from peft import PeftModel

base_model_name = "Qwen/Qwen2.5-3B-Instruct"
base_model = AutoModelForCausalLM.from_pretrained(base_model_name)

peft_model_path = "/home/lishuo1/efficient_reasoning/run/results_SFT/checkpoint-3750"
peft_model = PeftModel.from_pretrained(base_model, peft_model_path)

merged_model = peft_model.merge_and_unload()
output_dir = "/home/lishuo1/efficient_reasoning/run/results_SFT/checkpoint-3750"
merged_model.save_pretrained(output_dir)
```

```python
llm = LLM(model="/home/lishuo1/efficient_reasoning/run/results_ASFT/checkpoint-2814")

prompts = [
    "What is the capital of France?",
    "What is the capital of Germany?",
]
```
